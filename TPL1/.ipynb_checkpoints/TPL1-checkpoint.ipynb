### COMPARACIÓN TRANSFERENCIA NORMALIZADA
Según los cálculos, la trasnferencia de cheby normalizada nos quedó:
\begin{equation}
T(s) = \frac{0,71569}{(s')^3+(s')^2 * 1,252913 +s'*1,5349+0,71569} 
\end{equation}

Podemos corroborar si es correcto utilizando la función cheby1

```python
#!/usr/bin/env python3
# -*- coding: utf-8 -*-
"""
Created on Wed May 28 11:18:52 2025

@author: facu
"""

#-------------------------
# IMPORTACIÓN DE MÓDULOS Y CONFIGURACIONES PREVIAS
#------------------------
# Módulos externos

import numpy as np
import matplotlib as mpl
from matplotlib import pyplot as plt

# módulo de SciPy
from scipy import signal as sig

# Esta parte de código la agregamos SOLO en los notebooks para fijar el estilo de los gráficos.
fig_sz_x = 13
fig_sz_y = 7
fig_dpi = 80 # dpi
fig_font_size = 13

mpl.rcParams['figure.figsize'] = (fig_sz_x, fig_sz_y)
mpl.rcParams['figure.dpi'] = fig_dpi
plt.rcParams.update({'font.size':fig_font_size})

from pytc2.sistemas_lineales import analyze_sys, bodePlot, pzmap, pretty_print_bicuad_omegayq, pretty_print_SOS
from pytc2.general import print_latex, a_equal_b_latex_s

import matplotlib.patches as patches
from matplotlib.path import Path

def addCuadrado (axe_hdl,att_max=1, att_min=10, wc=1, ws=3,t_min_bottom=-100, t_max_top=100,wi=0,wf=0,i=0,color='red',trama='/',fill=False):
    #wi=winicio
    #wf=wfin
    if wi==0:
        wi=wc/2
    if wf==0:
        wf=ws*2
    #Tmin es determinada por att_max y viceversa
    t_min=1-att_max
    t_max=1-att_min
    
    vert_t_min =[
        (wi, t_min_bottom),  # left, bottom
        (wi, t_min),  # left, top
        (wc, t_min),  # right, top
        (wc, t_min_bottom),  # right, bottom
        (0., 0.),  # ignored
    ]
    vert_t_max =[
        (ws, t_max),  # left, bottom
        (ws, t_max_top),  # left, top
        (wf, t_max_top),  # right, top
        (wf, t_max),  # right, bottom
        (0., 0.),  # ignored
    ]
 
    codes = [Path.MOVETO, Path.LINETO, Path.LINETO, Path.LINETO, Path.CLOSEPOLY]
    
    path_t_min = Path(vert_t_min, codes)
    path_t_max = Path(vert_t_max, codes)
    patch_t_min = patches.PathPatch(path_t_min, color='red', hatch = '/', lw=1, fill=False)
    patch_t_max = patches.PathPatch(path_t_max, color='red', hatch = '/', lw=1, fill=False)
    
    axe_hdl.add_patch(patch_t_min)
    axe_hdl.add_patch(patch_t_max)
    
        

#----------------------
# INICIO DEL CÓDIGO
#--------------------
n=3
att_min=16
att_max=0.5
wc=1
ws=3
H_C=sig.cheby1(n, att_max, wc, btype='lowpass', analog=True, output='sos', fs=None)
H_C_ba=sig.cheby1(n, att_max, wc, btype='lowpass', analog=True, output='ba', fs=None)

#print(H_C)
"""
num_HC6=np.array(H_C[0])
num_HC6.resize(6)
#print(num_HC6) 
den_HC6=np.array(H_C[1])
den_HC6.resize(6)

for n in range(0,6):
    if n<len(H_C[0]):
        num_HC6[5-n] = num_HC6[len(H_C[0])-1-n]
    else:
        num_HC6[5-n] = 0
     
for n in range(0,6):
    if n<len(H_C[1]):
        den_HC6[5-n] = den_HC6[len(H_C[1])-1-n]
    else:
        den_HC6[5-n] = 0

#print(len(H_C[1]))
#print(H_C[0])
#print(num_HC6)
#print(den_HC6)

HC6= np.array(num_HC6)
HC6.resize(2,6)
np.put(HC6,[6,7,8,9,10,11],den_HC6)

#print(np.array(HC6))
"""
"""
_,axes_hdl_0=bodePlot(H_C, fig_id="1", filter_description="|T(S)| normalizado")
addCuadrado(axes_hdl_0[0],att_max=0.5, att_min=16,wc=1,ws=3,wi=1e-1, wf=3)
_, axes_hdl_02=bodePlot(H_C, fig_id="2", filter_description="|T(S)|")
axes_hdl_02[0].set_xlim(left=1e-1, right=1.1)
axes_hdl_02[0].set_ylim(top=1, bottom=-4)
addCuadrado(axes_hdl_02[0],att_max=0.5, att_min=16,wc=1,ws=3,wi=1e-1, wf=3)
"""

pretty_print_SOS(H_C)

```

## COMPARACIÓN TRANSFERENCIA DESNORMALIZADA
Ahora, desnormalizamos con las mismas normas que en el planteo teórico

```python
wn=3000*np.pi
wn_pot = [0,1,2,0,1,2]
for i in range(0,6):
    wn_pot[i]=np.power(wn,wn_pot[i])
wn_pot = np.array(wn_pot)

i = int(len(H_C)/2 + 1)
for n in range(0,i):
    for m in range (0,6):
        #print("%d , %d" % (n, m))
        #print("%f \n %f" % (H_C[n][m],wn_pot[m])) 
        #print("--------------")
        H_C[n][m]=H_C[n][m] * wn_pot[m]
        #print(H_C[n][m])
        #print("...........................")
pretty_print_SOS(H_C)  
```

Nuevamente, si se reordenaran los valores del numeradorobtendríamos los correspondientes al hallado en el pdf.
Procedemos graficando para corroborar si se cumple lo deseado.


```python
wc=2*np.pi*1.5e3
ws=2*np.pi*4.5e3
_,axes_hdl_1=bodePlot(H_C, fig_id="1", filter_description="|T(S)|")

_, axes_hdl_2=bodePlot(H_C, fig_id="2", filter_description="|T(S)|")

H_C_pzmap = sig.TransferFunction(H_C_ba[0], H_C_ba[1])
_,_,=pzmap(H_C_pzmap, fig_id="3")

addCuadrado(axes_hdl_1[0],att_max=0.5, att_min=16,wc=wc,ws=ws,wi=1e3, wf=1e5)
addCuadrado(axes_hdl_2[0],att_max=0.5, att_min=16,wc=wc,ws=ws,wi=1e3,wf=1e5)
axes_hdl_2[0].set_xlim(left=5e3, right=2e4)
axes_hdl_2[0].set_ylim(top=1, bottom=-4);
```

## FUNCIÓN TRASNFERENCIA CON VALORES
Con los valores de los componentes y las funciones transferencias de los circuitos propuestos, verificamos si se cumple lo deseado

```python
#------ VALORES
R1=R3=R4=18e3
R2=R5=33e3
C1=C2=5.6e-9

G1=1/R1
G2=1/R2
G3=1/R3
G4=1/R4
G5=1/R5


num_H1= G1/G3*np.array([0., 0., (G3/C1)**2])
den_H1= np.array([1.,G2/C1,(G3/C1)**2])
num_H2= np.array([0., 0., G5/C2])
den_H2= np.array([0., 1., G5/C2])


num_H = np.polymul(num_H1, num_H2)
den_H = np.polymul(den_H1, den_H2)

H1 = sig.TransferFunction(num_H1, den_H1)
H2 = sig.TransferFunction(num_H2, den_H2)
H = sig.TransferFunction(num_H, den_H)

_, axes_hdl_4=bodePlot(H1, fig_id="4", filter_description="|T1(S)|")
_,_,=bodePlot(H2, fig_id="4", filter_description="|T2(S)|")
_,_,=bodePlot(H, fig_id="4", filter_description="|T(S)|")

H_line = axes_hdl_4[0].get_lines()
H_line[0].set_linestyle('-.')
H_line[1].set_linestyle('--')
```
