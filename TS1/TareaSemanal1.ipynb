```python
import sympy as sp
from pytc2.remociones import remover_polo_dc
from pytc2.general import a_equal_b_latex_s, print_latex, s, symbfunc2tf, factorSOS

# tensiones y corrientes que usaremos para nuestros análisis
I1, I2, V1, V2, VX= sp.symbols("I1, I2, V1, V2, VX")
R1, R2, R3, YC= sp.symbols("R1, R2, R3, YC")
Q, wt, w0, C = sp.symbols("Q, wt, w0, C", real = True, posistive = True) 
```

```python
# modelo ideal negativamente realimentado
aa = sp.solve([ 
                VX*(1/R1 + 1/R2) - V1*(1/R1) - V2*(1/R2), 
                VX*(1/R3 + YC) - V1*(YC)
                ], 
                [V1, V2])

AV = aa[V2]/aa[V1] 
AV = sp.simplify(sp.expand(AV.subs(YC,s*C)))
#AV,_,_=factorSOS(AV)
print_latex(a_equal_b_latex_s('\\frac{V_2}{V_1}', AV))
```

Efectivamente, si realizo la distributiva con R1 en el denominador, llego a la misma expresión que hallé en el ejercicio 1.
Siguiendo con la resolución, si reemplazo la expresión por los valores normalizados llego a que...

```python
AV = sp.simplify(sp.expand(AV.subs(C*R3,1)))
print_latex(a_equal_b_latex_s('\\frac{V_2}{V_1}', AV))
```

Donde, nuevamente, si saco factor común R1 en el numerador y lo simplifico con el R1 del denominador, llego a la expresión hallada en el ejercicio 2
